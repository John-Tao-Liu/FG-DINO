```python
# import nltk
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')
# nltk.download('punkt_tab')
# nltk.download('averaged_perceptron_tagger_eng')
# nltk.download('stopwords')

import nltk
import os

# 1. 定义autodl-fs下的NLTK存储路径（按需调整）
nltk_dir = "/root/autodl-fs/demo/FG-DINO/nltk_data"
os.makedirs(nltk_dir, exist_ok=True)  # 确保目录存在
nltk.data.path.append(nltk_dir)  # 让NLTK优先读取该目录

# 2. 定义需要下载的包列表
required_packages = [
    'punkt', 'punkt_tab', 
    'averaged_perceptron_tagger', 'averaged_perceptron_tagger_eng',
    'stopwords'
]

# 3. 检查并下载（仅缺失时下载）
for pkg in required_packages:
    try:
        nltk.data.find(pkg)  # 检查是否已存在
        print(f"✅ {pkg} 已存在，跳过下载")
    except LookupError:
        print(f"📥 正在下载 {pkg} 到 {nltk_dir}")
        nltk.download(pkg, download_dir=nltk_dir, quiet=True)

print("✅ NLTK数据准备完成")
```
